Difference between interpreter and compiler:

    Compilers and interpreters are two different ways to translate a program from a high-level programming language to machine language.
    -> A compiler takes in the entire program and requires a lot of time to analyze the source code, while an interpreter takes a single line of code and very little time to analyze it.
    -> Compiled code runs faster than interpreted code, but interpreted code is easier to debug.
    -> A compiler displays all errors after compilation, while an interpreter displays errors of each line one by one.
    -> A compiler generates a stand-alone program, while an interpreter does not.
    -> A compiler is based on the translation linking-loading model, while an interpreter is based on the interpretation method.

What is Data Structure, why we need one in programs? 

    A data structure is a specialised format to organising, processing, retirving and storing datas.
    Data structure provides a way to organize and store data efficiently. Different types of data structures are suited to different kinds of applications, and choosing an appropriate data structure can greatly improve the performance of a program. For example, a program that needs to search through a large amount of data may use a data structure like a hash table, which allows for fast search times. Similarly, a program that needs to frequently add and remove items from a list may use a data structure like a linked list, which allows for fast insertions and deletions.
